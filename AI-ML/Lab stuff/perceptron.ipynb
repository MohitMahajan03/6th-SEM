Apply perceptron training rule on logic AND and OR gate 

```python
#AND Gate perceptron
def activation(out, threshold):
    if out >= threshold:
        return 1
    else:
        return 0

def perceptron(and_input):
    a = [0,0,1,1]
    b = [0,1,0,1]
    y = [0,0,0,1]
    w = [1.4,1.5]
    threshold = 1
    learning_rate = 0.1
    i=0
    print("Perceptron Training: ")
    print("######################")
    print("-----------------------")
    while i < 4:
        summation = a[i]*w[0] + b[i]*w[1]
        o = activation(summation, threshold)
        print("Input : ", str(a[i]) + " , " + str(b[i]))
        print("Weights : ", str(w[0]) + " , " + str(w[1]))
        print("Summation : ", str(summation) + " Threshold " + str(threshold))
        print("Actual output : ", str(y[i]) + " Predicated output " + str(o))
        if(o != y[i]):
            print("_______\nUpdating Weights")
            w[0] = w[0] + learning_rate*(y[i]-o)*a[i]
            w[1] = w[1] + learning_rate*(y[i]-o)*b[i]
            print("Updated Weights : ", str(w[0]) + " , "+ str(w[1]))
            i = -1
            print("Weights Updated Training Again:  ")
            print("#################################")
        i = i+1
        print("------------------------------")
    summation = and_input[0] * w[0] + and_input[1] * w[1]
    return activation(summation, threshold)

and_input = [1,1]
print("AND Gate Output For ", str(and_input) + " : " + str(perceptron(and_input)))
```

```python
#OR Gate Perceptron
def activation(out, threshold):
    if out >= threshold:
        return 1
    else:
        return 0

def perceptron(or_input):
    a = [0,0,1,1]
    b = [0,1,0,1]
    y = [0,1,1,1]
    w = [0.0,0.3]
    threshold = 0.4
    learning_rate = 0.5
    i=0
    print("Perceptron Training: ")
    print("######################")
    print("-----------------------")
    while i < 4:
        summation = a[i]*w[0] + b[i]*w[1]
        o = activation(summation, threshold)
        print("Input : ", str(a[i]) + " , " + str(b[i]))
        print("Weights : ", str(w[0]) + " , " + str(w[1]))
        print("Summation : ", str(summation) + " Threshold " + str(threshold))
        print("Actual output : ", str(y[i]) + " Predicated output " + str(o))
        if(o != y[i]):
            print("_______\nUpdating Weights")
            w[0] = w[0] + learning_rate*(y[i]-o)*a[i]
            w[1] = w[1] + learning_rate*(y[i]-o)*b[i]
            print("Updated Weights : ", str(w[0]) + " , "+ str(w[1]))
            i = -1
            print("Weights Updated Training Again:  ")
            print("#################################")
        i = i+1
        print("------------------------------")
    summation = or_input[0] * w[0] + or_input[1] * w[1]
    return activation(summation, threshold)

or_input = [0,0]
print("OR Gate Output For ", str(or_input) + " : " + str(perceptron(or_input)))
```

```python
import numpy as np
def threshold_activation_function(x, threshold = 0):
    return 1 if x > threshold else 0

def predict(inputs, weights, bias):
    weighted_sum = np.dot(inputs, weights) + bias
    
    return threshold_activation_function(weighted_sum)

def calculate_error(prediction, label):
    return label - prediction

def train_perceptron(training_inputs, labels, learning_rate = 0.01, num_epochs = 100):
    num_inputs = len(training_inputs[0])
    weights = np.random.rand(num_inputs)
    bias = np.random.rand()

    for epoch in range(num_epochs):
        for inputs, label in zip(training_inputs, labels):
            prediction = predict(inputs, weights, bias)

            error = calculate_error(prediction, label)

            weights += learning_rate* error * inputs
            bias += learning_rate * error
    return weights, bias

def calculate_accuracy(predections, labels):
    return np.mean(predections == labels)

if __name__ == "__main__":
    training_inputs = np.array([[0,0],[0,1],[1,0],[1,1]])
    labels = np.array([0,0,0,1])

    weights , bias = train_perceptron(training_inputs, labels)

    test_inputs = np.array([[0,0],[0,1],[1,0],[1,1]])
    predictions = [predict(inputs, weights, bias) for inputs in test_inputs]
    for inputs, prediction in zip(test_inputs, predictions):
        print(f"Input: {inputs}, Predicted Output: {prediction}")
    
    accuracy = calculate_accuracy(predictions, labels)
    print("Accuracy: ", accuracy)
    
```
